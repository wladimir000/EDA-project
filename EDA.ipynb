1. Do houses with higher grades have higher prices?

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    data=df,
    x="grade",
    y="price"
)

plt.title("Hauspreise nach Grade")
plt.xlabel("Grade")
plt.ylabel("Preis in USD")

plt.show()
```

```python
df.groupby("grade")["price"].mean().sort_index()
```

```python
df["grade"].corr(df["price"])
```

TRUE

Houses with higher grades tend to have significantly higher prices.

This is supported by:
- increasing median prices in the boxplot
- steadily rising average prices across grades
- a strong positive correlation (0.67) between grade and price

2. Have specific zip codes have higher prices?

```python
zipcode_prices = df.groupby("zipcode")["price"].mean().sort_values(ascending=False)

zipcode_prices.head(10)
```

```python
top_zipcodes = zipcode_prices.head(5).index

filtered_df = df[df["zipcode"].isin(top_zipcodes)]

plt.figure(figsize=(12,5))

sns.boxplot(
    data=filtered_df,
    x="zipcode",
    y="price"
)

plt.title("Preise in den teuersten Zipcodes")
plt.xlabel("Zipcode")
plt.ylabel("Preis in USD")

plt.show()
```

```python
zipcode_prices.plot(kind="bar", figsize=(15,5))
plt.title("Average House Price by Zipcode")
plt.ylabel("Average Price")
plt.show()
```

TRUE

Certain zipcodes show significantly higher average house prices than others.

This is supported by:
- noticeable differences in average prices across zipcodes
- higher median prices in the most expensive zipcodes
- clear regional price variation shown in the bar chart

3. Do houses with better views have higher prices?

```python
plt.figure(figsize=(10,5))

sns.boxplot(
    data=df,
    x="view",
    y="price"
)

plt.title("Hauspreise nach Aussicht")
plt.xlabel("View Rating")
plt.ylabel("Preis in USD")

plt.show()
```

```python
df.groupby("view")["price"].agg(["mean", "median", "count"])
```

TRUE

Houses with better view ratings tend to have higher prices.

This is supported by:
- increasing median prices in the boxplot
- higher average prices for higher view ratings
- an overall upward trend across all view categories

4. Do houses with larger living areas have higher prices?

```python
sns.regplot(
    data=df,
    x="sqft_living",
    y="price",
    scatter_kws={"alpha":0.4}
)
```

```python
df["sqft_living"].corr(df["price"])
```

TRUE

Houses with larger living areas generally tend to have higher prices.

This is supported by:
- a clear upward trend in the scatterplot
- a strong positive correlation between living area and price
- increasing prices as square footage increases
